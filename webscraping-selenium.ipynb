```python
urls=[]
for i in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']"):
    urls.append(i.get_attribute("href"))
urls
```

```python
from selenium.common.exceptions import NoSuchElementException
import time
job_title=[]
job_descr=[]
company=[]
location=[]
for i in urls[:10]:
    driver.get(i)
    time.sleep(3)
    try:
        job=driver.find_element_by_xpath("//h1[@class='jd-header-title']")
        job_title.append(job.text)
    except NoSuchElementException:
        job_title.append('-')
    try:
        job_des=driver.find_element_by_xpath("//section[@class='job-desc']")
        job_descr.append(job_des.text)
    except NoSuchElementException:
        job_descr.append('-')
    try:
        com=driver.find_element_by_xpath("//div[@class='jd-header-comp-name']")
        company.append(com.text)
    except NoSuchElementException:
        company.append('-')
    try:
        loc=driver.find_element_by_xpath("//div[@class='loc']")
        location.append(loc.text)
    except NoSuchElementException:
        location.append('-')
```

```python
Data_scientist_job=pd.DataFrame({})
Data_scientist_job['Job_title']=job_title
Data_scientist_job['Job_Location']=location
Data_scientist_job['Company']=company
Data_scientist_job['Description']=job_descr
Data_scientist_job
```
