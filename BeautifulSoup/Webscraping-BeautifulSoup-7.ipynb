```python
page1=requests.get("https://www.flipkart.com/search?q=mobile+under+20000rs&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_3_18_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_3_18_na_na_na&as-pos=3&as-type=RECENT&suggestionId=mobile+under+20000rs&requestId=f10d9b25-b0c4-453d-b007-b79bc806f5d5&as-searchtext=mobile%20under%2020000")
page1
```

```python
soup=BeautifulSoup(page1.content)
```

```python
print(soup.prettify())
```

```python
Data=soup.find_all('div',class_="_1AtVbE col-12-12")
Data
```

```python
brand=[]
price=[]
img=[]
rating=[]

a=soup.find_all('div',class_='_2kHMtA')
for x in a:
    brand.append(x.find('div',class_='_4rR01T').text.strip())    
print(brand)
for p in a:
    y=p.find('div',class_="_30jeq3 _1_WHN1")
    price.append(y.text.strip())
print(price)
    
for r in a:
    w=r.find('span',class_="_2_R_DZ")
    rating.append(w.text.replace('\xa0&\xa0'," ").strip())
    
print(rating)
```
