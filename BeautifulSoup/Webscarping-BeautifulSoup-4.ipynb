```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
page=requests.get("https://bookpage.com/reviews")
page
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
#Names of the Book
Names=soup.find_all('h4',class_="italic")
Names
```

```python
book_names=[]
for i in Names: 
    book_names.append(i.text.replace('\n',''))
book_names
```

```python
writer=soup.find_all('p',class_= "sans bold")
writer
```

```python
Author=[]
for i in writer: 
    Author.append(i.text.replace('\n',''))
Author
```

```python
genre=soup.find_all('p',class_= "genre-links hidden-phone")
genre
```

```python
book_genre=[]
for i in genre: 
    book_genre.append(i.text.replace('\n',''))
book_genre
```

```python
review=soup.find_all('p',class_= "excerpt")
review
```

```python
book_review=[]
for i in review: 
    book_review.append(i.text.replace('\n',''))
book_review
```

```python
#Checking the length
print('book_names:',len(Names))
print('book_genre:',len(genre))
print('Author:',len(writer))
print('book_review:',len(review))
```

```python
Books=pd.DataFrame({})
Books['book_names']=book_names
Books['Author']=Author
Books['genre']=book_genre
Books['review']=book_review
Books
```
