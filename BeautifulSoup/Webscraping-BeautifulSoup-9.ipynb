```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
page=requests.get("https://internshala.com/fresher-jobs")
page
```

```python
soup=BeautifulSoup(page.content,'html.parser')
print(soup.prettify())
```

```python
data=soup.find_all('div',class_="internship_meta")
data
```

```python
title=[]
company=[]


for i in data:
    x=i.find('div',class_="heading_4_5 profile")
    title.append(x.text.replace('\n',""))
title

```

```python
for i in data:
    y=i.find('div',class_="heading_6 company_name")
    company.append(y.text.strip())
company
```

```python
Data=pd.DataFrame({})
Data['title']=title
Data['company']=company
Data
```
